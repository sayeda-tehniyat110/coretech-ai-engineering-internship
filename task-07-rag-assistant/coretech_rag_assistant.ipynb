```python
# Task 07: CoreTech Innovation RAG-Style Knowledge Assistant Ecosystem
!pip install -q sentence-transformers

import numpy as np
import torch
from sentence_transformers import SentenceTransformer

# 1. Defining the 5 Internal Knowledge Base Documents
knowledge_base = {
    "Company Profile": (
        "CoreTech Innovation is an advanced engineering firm founded by Computer Systems Engineering students. "
        "Based in Sindh, Pakistan, the company specializes in bridging raw backend engineering, hardware interfaces, "
        "and microprocessor logic with modern intelligent software networks and high-end digital branding environments."
    ),
    "Services": (
        "CoreTech Innovation provides four core enterprise deliverables: "
        "1. AI Automated Solutions (custom NLP models, automated workflows, semantic search integration). "
        "2. Full-Stack Web Development (robust backend architecture using PHP/SQL paired with highly responsive frontends). "
        "3. UI/UX System Design (Figma wireframing, fluid application prototypes). "
        "4. Graphic Design & Visual Branding (premium logo suites and dynamic social media content)."
    ),
    "Project Process": (
        "Our rigorous engineering lifecycle operates in clear developmental phases: "
        "Phase 1: Discovery Session (scoping client requirements and infrastructure analysis). "
        "Phase 2: UI/UX Prototyping (designing layouts in Figma for client review). "
        "Phase 3: Core Functional Sprint (writing code and databases, committing daily updates into transparent GitHub branches). "
        "Phase 4: Triage & Deployment (quality testing, production deployment, and complete repository asset handover)."
    ),
    "Pricing Sample": (
        "CoreTech Innovation calculates pricing dynamically based on project architectural complexity and developmental hours. "
        "A baseline Full-Stack Web Application combined with a custom rule-based Chatbot script starts at a baseline sample investment of $1,500 USD. "
        "All projects require custom scope analysis, and clients receive complete code ownership upon milestone clearance."
    ),
    "FAQs": (
        "Frequently Asked Questions: "
        "Q1: Who owns the final source code? A: The client receives absolute IP and ownership upon project completion via GitHub. "
        "Q2: How do we track project progression? A: Clients get direct access to dedicated GitHub repositories to audit real-time code commits. "
        "Q3: Do you support legacy database systems? A: Yes, we provide full migration support for custom SQL setups."
    )
}

print("CoreTech 5 Corporate Knowledge Documents initialized successfully!")
```

```python
# 2. Loading the Semantic Embedding Model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Extract document names and their text content
doc_titles = list(knowledge_base.keys())
doc_texts = list(knowledge_base.values())

# Compute dense vector embeddings for our corporate corpus
corpus_embeddings = embedding_model.encode(doc_texts, convert_to_tensor=True)

print(f"Successfully vectorized {len(doc_titles)} documents using 384-dimensional space.")
```

```python
# 3. RAG Execution Pipeline Function
def coretech_rag_assistant(user_question):

    query_embedding = embedding_model.encode(user_question, convert_to_tensor=True)
    similarity_scores = torch.nn.functional.cosine_similarity(query_embedding, corpus_embeddings)

    best_match_idx = torch.argmax(similarity_scores).item()
    highest_score = similarity_scores[best_match_idx].item()

    retrieved_source = doc_titles[best_match_idx]
    retrieved_context = doc_texts[best_match_idx]
    if highest_score < 0.25:
        generated_answer = "I'm sorry, I couldn't find specific details regarding that query in the CoreTech knowledge base. Please contact our human operations team."
        retrieved_source = "None (Insufficient Confidence)"
    else:
        generated_answer = f"Based on our internal corporate records, {retrieved_context}"
    print(f"\n[USER QUESTION]: {user_question}")
    print(f"[RETRIEVED SOURCE]: {retrieved_source} (Confidence: {highest_score:.4f})")
    print("-" * 60)
    print(f"[GENERATED RESPONSE]:\n{generated_answer}\n")
    print("=" * 60)

print("RAG System Pipeline compiled and ready for testing!")
```

```python
# 4. Testing the Assistant with diverse client search query scenarios
print("EXECUTING CORETECH RAG KNOWLEDGE ASSISTANT RUN:")
print("=" * 60)
coretech_rag_assistant("What kind of artificial intelligence or front-end work do you do?")
coretech_rag_assistant("How can I track my project code and commits?")
coretech_rag_assistant("What is the starting price for a full-stack web application?")
```
