```python
%%writefile app_capstone.py
# Task 12: CoreTech Capstone Innovation - Advanced RAG & Semantic Search Suite (Contrast Fix)
import streamlit as st
import pandas as pd
import numpy as np
import io
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Production UI Frame Configuration
st.set_page_config(
    page_title="CoreTech Semantic RAG Assistant",
    page_icon="🧠",
    layout="wide"
)

# Premium Midnight Violet Corporate Aesthetic Theme Injection (Contrast & Brightness Fixed)
st.markdown("""
    <style>
    /* Vibrant Midnight Violet background update */
    .main, [data-testid="stAppViewContainer"], [data-testid="stHeader"] {
        background-color: #131026 !important;
        color: #f8f9fa !important;
    }

    /* Strong bright white headings for high contrast */
    h1, h2, h3 {
        color: #d1c4e9 !important;
        font-weight: 700 !important;
        text-shadow: 0px 2px 4px rgba(0,0,0,0.5);
    }

    /* Enhanced text legibility rules */
    p, span, label, div {
        color: #f1f1f1 !important;
        font-size: 16px !important;
    }

    /* Input box crisp alignment mapping */
    .stTextInput>div>div>input, textarea, [data-testid="stChatInputTextArea"] {
        background-color: #211c3d !important;
        color: #ffffff !important;
        border: 2px solid #673ab7 !important;
        border-radius: 8px !important;
    }

    /* Information panel card background alignment */
    div.stAlert {
        background-color: #1c183a !important;
        border: 1px solid #7e57c2 !important;
        border-radius: 10px !important;
    }
    </style>
    """,
    unsafe_allow_html=True
)

# 2. Hardcoded 30-Record Structured Corporate Knowledge Base
knowledge_base_csv = """Topic,Context
"AI Automation","CoreTech deploys end-to-end AI Automation pipelines using Natural Language Processing (NLP) models to parse corporate communication streams."
"AI Lead Scoring","Our automated algorithmic lead scoring engine models customer engagement parameters company scale and budget signals across 0-100 metrics."
"Semantic Search","We engineer semantic search setups mapping query text vectors into mathematical vector space using TF-IDF token matrices and Cosine Similarity."
"Chatbot Architecture","CoreTech customer assistants use multi-tier fallback arrays tracking structural user intents before routing to technical support mail grids."
"Web Development Stack","Our full-stack engineering grid relies on robust production PHP backends mapped to optimized relational SQL database infrastructure."
"Frontend Frameworks","We engineer highly responsive fluid user interfaces using semantic vanilla HTML5 elements structural CSS rules and modular JavaScript."
"UI/UX Design Platform","CoreTech utilizes Figma for industry standard high-fidelity wireframing component design layout configuration and prototyping systems."
"Graphic Asset Creation","We design premium visual branding assets using Adobe Illustrator Photoshop and Canva for rapid asset layout orchestration."
"Logo & Brand Identity","Our creative division delivers complete custom corporate identities logo design sheets typographical frameworks and color asset guidelines."
"Social Media Visuals","We design beige social media design elements vector marketing banners and interactive visual layouts for corporate branding campaigns."
"Project Code Repositories","All client source code files are managed inside dedicated private GitHub repositories providing absolute real-time commit auditing access."
"Agile Sprint Delivery","Developmental schedules operate on rigid weekly agile sprints with transparent milestone updates logged directly inside client repository logs."
"IP & Source Ownership","Clients receive complete intellectual property legal compliance frameworks and absolute source code ownership upon final sprint milestone clearance."
"Project Pricing Matrix","Engineering costs are calculated dynamically based on developmental workload hours system architectural layer depth and resource requirements."
"Baseline Website Package","Standard full-stack web applications paired with baseline intent parsing chatbot engines start at a entry investment of 1500 USD."
"Enterprise SQL Engineering","Our engineers build custom relational schema architectures query performance optimizations indexing strategies and data backup pipelines."
"Data Migration Support","CoreTech provides complete safe data migration pipelines shifting legacy infrastructure models into scalable optimized SQL instances."
"Server Cloud Hosting","We guide enterprise clients through production deployment server network configurations cloud infrastructure routing and secure hosting systems."
"Post-Launch Warranty","Every software system launched includes an explicit 30-day post-deployment warranty phase to intercept and correct unexpected server bugs completely free."
"Technical Maintenance Grid","We offer optional structured monthly technical support packages covering runtime database security evaluation and dependency monitoring."
"Change Request Protocol","Modifications requested outside the verified project statement parameters are scheduled for phase two launches or treated as paid extensions."
"Operational Response SLA","Our dedicated technical operations management team reviews inbound inquiry forms and guarantees formal project feedback within 24 business hours."
"Geographic Operations Hub","CoreTech Innovation operates its primary corporate engineering network and developmental workspace from Sindh Pakistan."
"Founding Academic Core","The agency was founded by a specialized group of Bachelor of Engineering in Computer Systems Engineering students focused on intelligent systems."
"Academic Competency Grid","Our core engineering foundation comprises deep mastery in hardware systems logic design network management microprocessor architectures and full-stack software development."
"Remote Internship Tracks","CoreTech frequently organizes structured remote technical internships allowing aspiring software developers to contribute to active production pipelines."
"Client Communication Channels","Official project discussions are routed cleanly through secure Slack workspaces formal email chains and recorded technical review calls."
"System Performance Scaling","We enforce rigid code reviews optimization patterns and network query caching architectures to ensure software performance remains fast under scale."
"Database Security Measures","Relational database instances are secured using encrypted password configurations parameterized database queries and regular automated snapshots."
"Data Analytics Pipelines","CoreTech deploys python analytical models data extraction script structures and automated report registries to optimize corporate insight management."
"""

df_kb = pd.read_csv(io.StringIO(knowledge_base_csv))

# 3. Vectorization and RAG Retrieval System Math Matrix Engine
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df_kb['Context'])

def perform_semantic_rag(user_query):
    query_vector = vectorizer.transform([user_query])
    similarity_scores = cosine_similarity(query_vector, tfidf_matrix).flatten()
    best_match_idx = np.argmax(similarity_scores)
    highest_score = similarity_scores[best_match_idx]

    if highest_score >= 0.15:
        topic = df_kb.iloc[best_match_idx]['Topic']
        context = df_kb.iloc[best_match_idx]['Context']
        synthesized_response = f"**[Verified Domain Context: {topic}]**\n\nAccording to CoreTech's enterprise knowledge registry: {context}"
        return synthesized_response, highest_score
    else:
        return "I am sorry, the system could not locate sufficient semantic overlap in the internal knowledge base for this query. Please route your request to our operations team at operations@coretech.com.", highest_score

# 4. Streamlit Front-End Interface Presentation Layout
st.markdown("<h1>🧠 CoreTech Capstone AI Suite</h1>", unsafe_allow_html=True)
st.markdown("### Production-Grade Semantic Retrieval (RAG) & Vector Engine")
st.write("This intelligence terminal matches natural language query inputs into high-dimensional vector spaces using a TF-IDF tokenization matrix and Cosine Similarity equations to retrieve absolute facts.")

st.markdown("---")

col1, col2 = st.columns([2, 1])

with col1:
    st.subheader("Interactive Inquiry Interface")
    user_prompt = st.text_input("Enter your engineering or system query here:", placeholder="e.g. How do you handle database security and backend stack?")

    if user_prompt:
        with st.spinner("Processing high-dimensional similarity equations..."):
            ai_response, score = perform_semantic_rag(user_prompt)

        st.markdown("### 🤖 Synthesized System Response:")
        st.info(ai_response)

with col2:
    st.subheader("Engine Telemetry Logs")
    if user_prompt:
        st.metric(label="Cosine Similarity Confidence Metric", value=f"{score:.4f}")
        if score >= 0.15:
            st.success("✅ Confidence Threshold Passed: Context Validated")
        else:
            st.error("❌ Fallback Triggered: Low Mathematical Overlap")
    else:
        st.write("Awaiting query telemetry calculation markers...")

st.markdown("---")
with st.expander("Audit Internal Knowledge Base Data Grid (30 Verified Rows)"):
    st.dataframe(df_kb, use_container_width=True)
```
