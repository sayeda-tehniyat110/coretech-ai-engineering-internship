```python
agency_name = "CoreTech"
founded_year = 2024
hourly_rate = 75.50
is_accepting_clients = True

print(f"Agency: {agency_name} (Type: {type(agency_name)})")
print(f"Founded: {founded_year} (Type: {type(founded_year)})")
print(f"Rate: ${hourly_rate}/hr (Type: {type(hourly_rate)})")
print(f"Active: {is_accepting_clients} (Type: {type(is_accepting_clients)})")
```

```python
coretech_services = [
    ("Web Development", 1200),
    ("Logo & Branding", 500),
    ("Social Media Design", 350),
    ("Software Troubleshooting", 150)
]

print("Available CoreTech Services")
for service, price in coretech_services:
    print(f"• {service:<25} | Starting from: ${price}")
```

```python
client_inquiry = {
    "client_name": "Syeda Tehniyat",
    "project_type": "Web Development",
    "budget": 1500,
    "timeline_weeks": 4,
    "requirements": ["Responsive design", "Contact form", "Admin panel"]
}

print("Current Client Inquiry")
for key, value in client_inquiry.items():
    print(f"{key.replace('_', ' ').title()}: {value}")
```

```python
def recommend_service(budget, project_type):
    """
    Checks if the client's budget matches the requested service.
    """
    service_prices = dict(coretech_services)

    if project_type not in service_prices:
        return f"Sorry, we currently do not offer '{project_type}' services."

    required_min = service_prices[project_type]

    if budget >= required_min:
        return f"Great news! Your budget of ${budget} covers the minimum cost for {project_type} (${required_min}). We can proceed with your project!"
    else:
        shortfall = required_min - budget
        return f"The minimum package for {project_type} starts at ${required_min}. You are ${shortfall} short of the starting price."
print(recommend_service(client_inquiry["budget"], client_inquiry["project_type"]))
```

```python
def coretech_assistant():
    print("Welcome to the CoreTech Interactive Assistant!")
    name = input("Please enter your name: ")
    print(f"\nHello {name}! Which service are you interested in?")
    service_list = [s[0] for s in coretech_services]
    for idx, service in enumerate(service_list, 1):
        print(f"{idx}. {service}")

    try:
        choice = int(input("\nEnter the service number: ")) - 1
        if 0 <= choice < len(service_list):
            selected_service = service_list[choice]
            client_budget = float(input("What is your estimated budget ($)? "))
            print("\nAssessment Result")
            print(recommend_service(client_budget, selected_service))
        else:
            print("Invalid selection. Please restart the assistant and choose a valid number.")
    except ValueError:
        print("Please enter valid numeric values for selection and budget.")
coretech_assistant()
```
