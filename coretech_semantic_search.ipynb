```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def coretech_semantic_search():
    try:
        kb_df = pd.read_csv("coretech_knowledge_base.csv")
    except FileNotFoundError:
        print("Oops! CSV file is missing. Run the first cell to create it.")
        return

    print("===================================")
    print("🤖 CoreTech Support Search Engine")
    print("===================================")

    user_query = input("Ask me anything about CoreTech services or tech stack: ").strip()

    if user_query == "":
        print("Please type a word or sentence to search.")
        return

    text_list = kb_df['Text_Content'].tolist()
    text_list.append(user_query)

    # Vectorization and matching logic
    vectorizer = TfidfVectorizer(stop_words='english')
    matrix = vectorizer.fit_transform(text_list)

    scores = cosine_similarity(matrix[-1], matrix[:-1])[0]
    kb_df['Match_Score'] = scores

    # Get top 3 matching rows
    top_results = kb_df.sort_values(by='Match_Score', ascending=False).head(3)

    print(f"\n My Top 3 Search Results for: '{user_query}' ")
    print("-" * 50)

    rank = 1
    for index, row in top_results.iterrows():
        percentage = round(row['Match_Score'] * 100, 2)
        print(f"Rank {rank} | Category: {row['Category']} | Similarity: {percentage}%")
        print(f"Text: {row['Text_Content']}")
        print("-" * 50)
        rank += 1

# Run the system
coretech_semantic_search()
```
