```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# My function for the Semantic Search system task
def coretech_semantic_search():
    # Load our 30 records csv file that we generated in step 1
    try:
        kb_df = pd.read_csv("coretech_knowledge_base.csv")
    except FileNotFoundError:
        print("Oops! CSV file is missing. Run the first cell to create it.")
        return

    print("===================================")
    print("🤖 CoreTech Support Search Engine")
    print("===================================")

    # Get what the user wants to search from the input box
    user_query = input("Ask me anything about CoreTech services or tech stack: ").strip()

    # Check if user just pressed enter without typing anything
    if user_query == "":
        print("Please type a word or sentence to search.")
        return

    # Convert our column to a python list so we can append the query
    text_list = kb_df['Text_Content'].tolist()
    text_list.append(user_query) # Putting query at the very end of the list

    # Using TF-IDF to turn words into numbers for the math matching
    # Added English stop words so common words like 'is', 'the', 'are' get ignored
    vectorizer = TfidfVectorizer(stop_words='english')
    matrix = vectorizer.fit_transform(text_list)

    # Calculate cosine similarity matrix
    # matrix[-1] is our user query (last item). We compare it with all rows before it matrix[:-1]
    scores = cosine_similarity(matrix[-1], matrix[:-1])[0]

    # Saving scores back into a new column in our dataframe to sort them
    kb_df['Match_Score'] = scores

    # Sorting from highest score to lowest score and getting top 3
    top_results = kb_df.sort_values(by='Match_Score', ascending=False).head(3)

    print(f"\n My Top 3 Search Results for: '{user_query}' ")


    # Loop to print out the final rank and data nicely
    rank = 1
    for index, row in top_results.iterrows():
        # Converting raw score decimal into a readable percentage
        percentage = round(row['Match_Score'] * 100, 2)

        print(f"Rank {rank} | Category: {row['Category']} | Similarity: {percentage}%")
        print(f"Text: {row['Text_Content']}")
        print("-" * 50)
        rank += 1

# Running my search function
coretech_semantic_search()
```
