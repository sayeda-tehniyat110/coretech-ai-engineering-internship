```python
# Task 08: CoreTech Inbound Client Inquiry ML Text Classification Framework (Fixed Dataset Count)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# 1. Defining a comprehensive dataset of 64 raw customer inquiry messages balanced across 8 categories
raw_data = [
    # Web Development (8 items)
    ("I need a responsive custom portfolio website built using PHP and an SQL database layout.", "Web Development"),
    ("Looking to build a responsive e-commerce clone with dynamic product grid displays and a login panel.", "Web Development"),
    ("Can you code a full-stack dashboard for our corporate website using standard CSS and JavaScript?", "Web Development"),
    ("We want a modern landing page for our business setup with dynamic portfolio elements.", "Web Development"),
    ("Need a backend software system optimized to handle high traffic using dynamic PHP script routines.", "Web Development"),
    ("Looking for a team to design and build a secure corporate profile website from scratch.", "Web Development"),
    ("Want an online automated web portal where clients can download invoice documentation files directly.", "Web Development"),
    ("Can you deploy a portfolio gallery platform featuring metadata indexing fields for files?", "Web Development"),

    # App Development (8 items)
    ("We need a custom mobile application developed for both Android and iOS devices.", "App Development"),
    ("Looking for cross-platform app developers to build an on-demand food delivery tracker application.", "App Development"),
    ("Can you program a native flutter application mapped out to match our internal inventory systems?", "App Development"),
    ("Need a secure mobile application setup with live user tracking modules and push notifications.", "App Development"),
    ("Want an enterprise mobile dashboard app that can fetch backend data from our main servers.", "App Development"),
    ("Looking to code an android app framework featuring material design components and payment gateway integrations.", "App Development"),
    ("Need a mobile tracking tool application built for our local transport operations business.", "App Development"),
    ("Can you create a cross platform mobile application layout tracking corporate delivery metrics?", "App Development"),

    # UI/UX Design (8 items)
    ("Looking for an expert designer to create high-fidelity responsive wireframes inside Figma.", "UI/UX Design"),
    ("We need interactive mobile app prototypes and fluid visual layouts mapped out in Figma panels.", "UI/UX Design"),
    ("Can your creative design team evaluate our website layouts and optimize user navigation routes?", "UI/UX Design"),
    ("Need a modern human-centered landing page mockup layout for our software startup system.", "UI/UX Design"),
    ("Want professional user interface prototypes for an online checkout page tracking visual flows.", "UI/UX Design"),
    ("Looking for a clean wireframe model for an enterprise database dashboard architecture project.", "UI/UX Design"),
    ("Can you design a fluid user experience journey prototype using premium structural canvas tools?", "UI/UX Design"),
    ("Need high fidelity component UI libraries mapped out for our engineering startup web platforms.", "UI/UX Design"),

    # Digital Marketing (8 items)
    ("We want to launch a targeted social media marketing ad campaign on LinkedIn and Facebook platforms.", "Digital Marketing"),
    ("Looking for a media agency strategist to scale our corporate branding visibility online.", "Digital Marketing"),
    ("Need specialized ad setup management and continuous budget spending metrics optimization solutions.", "Digital Marketing"),
    ("Can you manage our social media handles and write continuous conversion-focused promotional copy lines?", "Digital Marketing"),
    ("Want an expert to run automated lead generation ad workflows targeting high-value corporate accounts.", "Digital Marketing"),
    ("Looking to boost our business page engagement metrics via customized graphic media assets.", "Digital Marketing"),
    ("Need premium brand identity deployment tracking for our monthly product launch events.", "Digital Marketing"),
    ("We need an executive branding strategy package to boost our conversion metrics online.", "Digital Marketing"),

    # SEO (8 items)
    ("Our website traffic is dropping. We need a comprehensive search engine optimization structural audit.", "SEO"),
    ("Looking to optimize our blog pages and target competitive technical high-volume keywords.", "SEO"),
    ("Can you fix our technical search visibility bugs indexing structural tags and sitemaps?", "SEO"),
    ("Need organic search scaling models to rank our local business profile page higher on Google.", "SEO"),
    ("Want premium link-building strategies and off-page indexing optimization to boost our page rankings.", "SEO"),
    ("Looking for an automated tool layout keyword research suite setup for our content creators.", "SEO"),
    ("Can you optimize our corporate landing pages for search engines to boost inbound organic lead ratios?", "SEO"),
    ("Need deep keyword research mapping to outrank competitors on search result indexes globally.", "SEO"),

    # Software Solutions (8 items)
    ("We require an enterprise automated script framework to handle data synchronization across systems.", "Software Solutions"),
    ("Looking to build a custom desktop inventory management system using object-oriented C++ patterns.", "Software Solutions"),
    ("Can you develop an automated pipeline script tracking internal server errors and logs dynamically?", "Software Solutions"),
    ("Need a professional standalone quiz game platform engine developed with structured backend database files.", "Software Solutions"),
    ("Want a custom terminal system solver tool running advanced back-tracking algorithmic structures.", "Software Solutions"),
    ("Looking to build an enterprise resource planning software architecture for scalable startups.", "Software Solutions"),
    ("Can you deploy a standalone data analysis desktop script framework utilizing optimization parameters?", "Software Solutions"),
    ("We need a custom automated terminal program processing matrix transformations dynamically.", "Software Solutions"),

    # General Inquiry (8 items)
    ("Hi there! I wanted to check if CoreTech has open slots for remote technology student internships?", "General Inquiry"),
    ("Could you provide a copy of your technical services portfolio booklet and client reviews list?", "General Inquiry"),
    ("Hello, what are your team's standard operational business hours during corporate working days?", "General Inquiry"),
    ("I am an engineering student. Who should I reach out to regarding project collaboration titles?", "General Inquiry"),
    ("Could you share a sample pricing roadmap document mapping out standard structural timelines?", "General Inquiry"),
    ("Hello! Do you have custom support contact channels or telephone assistance routing available?", "General Inquiry"),
    ("I want to know if your team offers specialized corporate consulting consultation seminars?", "General Inquiry"),
    ("Can you explain how your student training programs coordinate over remote sprints?", "General Inquiry"),

    # Complaint (8 items)
    ("Our application server is down. Users are seeing severe runtime database timeout errors.", "Complaint"),
    ("The live checkout system crashed completely during payment processing steps this afternoon.", "Complaint"),
    ("The frontend layout buttons are distorted on mobile devices. This bug violates project criteria.", "Complaint"),
    ("Your support team has not replied to our critical project scope modification email for 48 hours.", "Complaint"),
    ("We noticed a major security flaw allowing unauthorized data entries into the SQL database view.", "Complaint"),
    ("The final code handover repository is missing necessary setup documentation files and installation guides.", "Complaint"),
    ("Our website loading speeds are extremely slow after your recent production code deployment push.", "Complaint"),
    ("The chatbot loop crashes every single time an out-of-vocabulary statement token is input.", "Complaint")
]

# 2. Construct flat DataFrame structure and export
df_inquiries = pd.DataFrame(raw_data, columns=['Message', 'Category'])
df_inquiries.to_csv('coretech_client_inquiries.csv', index=False)
print(f"1. 'coretech_client_inquiries.csv' generated with {len(df_inquiries)} calibrated rows.\n")


# --- STEP 3: ML MODEL TRAINING PIPELINE ---
X = df_inquiries['Message']
y = df_inquiries['Category']

# Split data into training and validation tiers
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# Convert text messages into dense features using TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english', sublinear_tf=True)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Initialize and train the Multinomial Naive Bayes Classifier model
inquiry_classifier = MultinomialNB(alpha=0.5)
inquiry_classifier.fit(X_train_tfidf, y_train)

# Run classification evaluations on the testing tier
predictions = inquiry_classifier.predict(X_test_tfidf)

# Statistical Metrics Summary Output
accuracy = accuracy_score(y_test, predictions)
print("="*70)
print(f"🤖 CORETECH INNOVATION AUTOMATED INBOUND INQUIRY ML CLASSIFIER")
print("="*70)
print(f"MODEL ACCURACY SCORE: {accuracy * 100:.2f}%\n")
print("[CLASSIFICATION METRICS REPORT]:")
print(classification_report(y_test, predictions, zero_division=0))
print("="*70)
```

```python
# 4. Live Predictive Engine Verification Wrap
def predict_incoming_ticket(custom_message):
    vectorized_input = vectorizer.transform([custom_message])
    predicted_label = inquiry_classifier.predict(vectorized_input)[0]
    print(f"Incoming Ticket : '{custom_message}'")
    print(f"AI Classification: ➔ [{predicted_label}]\n" + "-"*65)

print("RUNNING LIVE AI INFERENCE CHECKS:")
print("-"*65)
predict_incoming_ticket("Our database is down and throwing critical errors to live clients!")
predict_incoming_ticket("We want a modern high-fidelity layout wireframe crafted in Figma.")
```
