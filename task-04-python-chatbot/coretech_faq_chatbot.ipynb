```python
# Task 04: CoreTech FAQ Dataset & Foolproof Intent Routing Chatbot Pipeline (Fixed Formatting)
import pandas as pd
import io

# 1. Automatically Generating the Comprehensive 25 FAQ Dataset with Encased Quotes to prevent CSV Splitting
faq_data = """Question,Answer
"what services does coretech provide","CoreTech Innovation provides AI Automated Solutions, Full-Stack Web Development (PHP/SQL), UI/UX Optimization (Figma), and Graphic Branding."
"where is coretech located","Our core corporate engineering hub is based in Sindh, Pakistan."
"who founded coretech","CoreTech was founded by a team of Computer Systems Engineering students specializing in scalable software and intelligent networks."
"what is coretech background","Our team has a deep engineering foundation in microprocessor architecture, logic design, backend systems, and modern visual identities."
"how can i request a project quote","You can request a custom pricing quote by filling out our project inquiry form directly on the official CoreTech website."
"what is the starting price for websites","A baseline Full-Stack Web Application combined with a basic rule-based chatbot script starts at a baseline sample investment of $1500 USD."
"how do you calculate project pricing","Our pricing is calculated dynamically based on project architectural complexity and required developmental engineering hours."
"who owns the final source code","The client receives absolute IP and full code ownership upon project completion and final milestone clearance."
"how can i track my project progress","Clients get direct access to dedicated GitHub repositories to audit real-time source code commits and engineering sprint updates."
"what is your web development tech stack","We build robust backend systems using PHP and optimized SQL databases paired with highly responsive frontends using HTML, CSS, and JavaScript."
"do you design logos","Yes, we offer premium visual branding suites, corporate logo design, and customized brand identity creation."
"can you design social media content","Yes, we specialize in digital graphic design, visual assets, and creating social media layouts for corporate marketing campaigns."
"do you use figma for ui design","Yes, we utilize Figma for high-fidelity wireframing, fluid layout systems, and interactive user interface prototyping."
"do you use canva for visual assets","Yes, we use Canva and Adobe Illustrator for rapid text styles, social media vector assets, and layout graphics."
"what is an ai automated solution","It includes custom Natural Language Processing (NLP) models, automated lead triage workflows, and smart semantic search integrations."
"do you build chatbots for clients","Yes, we build custom rule-based and semantic-aware chatbot assistants using Python for customer support automation workflows."
"do you help with server hosting","Yes, we guide clients through cloud infrastructure management, production deployment, and server hosting configurations."
"what happens if i request extra features","Extra features outside the original project scope can be deployed in Phase 2 after launch or treated as a paid extension sprint."
"do you support legacy databases","Yes, we provide complete data migration, performance scaling, and optimization support for custom legacy SQL setups."
"what is your response time for inquiries","Our engineering operations team reviews all inbound project forms and responds back within 24 business hours."
"can university students join coretech","CoreTech is built on engineering talent and regularly announces remote internship tracks for technology students."
"do you offer monthly maintenance","Yes, we provide structured monthly technical maintenance packages for database security tracking and frontend monitoring."
"how do you handle project delays","We operate on strict agile weekly sprints logged transparently on GitHub to ensure milestones are met perfectly on schedule."
"what if i encounter an error after launch","We offer an explicit 30-day post-deployment warranty period to fix any unexpected deployment bugs completely free of cost."
"how do i contact support","You can drop an email to our official communication inbox or route your query instantly via our automated web assistant."
"""

# Convert the raw text data into a DataFrame and write out the local CSV file
df_faq = pd.read_csv(io.StringIO(faq_data), dtype=str)
df_faq.to_csv('coretech_faq.csv', index=False)
print("1. 'coretech_faq.csv' dataset successfully created with 25 complete row arrays.")


# --- 2. FOOLPROOF CORE-INTENT GATEWAY ENGINE ---
def run_coretech_keyword_chatbot():
    faq_df = pd.read_csv('coretech_faq.csv', dtype=str).fillna("")

    print("\n" + "="*75)
    print("🤖 CORETECH CLIENT ONBOARDING INTERACTIVE CHATBOT SYSTEM")
    print("Status: Operational | Database: coretech_faq.csv | Matching Type: Intent Scan")
    print("Type 'exit' or 'quit' at any point to close the communication logs.")
    print("="*75 + "\n")

    while True:
        # Normalize incoming raw strings
        user_query = input("You: ").strip().lower().replace('?', '').replace('.', '').replace(',', '')

        # Termination check
        if user_query in ['exit', 'quit']:
            print("\nBot: Thank you for communicating with CoreTech Innovation. Professional logs closed. Goodbye!")
            break

        if not user_query:
            print("Bot: Please enter a valid inquiry regarding our engineering or design services.\n")
            continue

        target_answer = None

        # Broad-Spectrum Intent Mapping Routing Control
        if any(w in user_query for w in ["service", "provide", "offer", "track", "do you do"]):
            target_answer = faq_df.iloc[0]['Answer']
        elif any(w in user_query for w in ["price", "cost", "investment", "rate", "fee", "charge"]):
            target_answer = faq_df.iloc[5]['Answer']
        elif any(w in user_query for w in ["where", "location", "located", "address", "hub", "pakistan"]):
            target_answer = faq_df.iloc[1]['Answer']
        elif any(w in user_query for w in ["found", "start", "background", "history", "who"]):
            target_answer = faq_df.iloc[2]['Answer']
        elif any(w in user_query for w in ["progress", "github", "commit", "source code", "repo"]):
            target_answer = faq_df.iloc[8]['Answer']
        elif any(w in user_query for w in ["figma", "ui", "ux", "wireframe", "prototype"]):
            target_answer = faq_df.iloc[12]['Answer']
        elif any(w in user_query for w in ["logo", "branding", "banner", "illustrator", "graphic"]):
            target_answer = faq_df.iloc[10]['Answer']
        elif any(w in user_query for w in ["intern", "internship", "student", "join", "career"]):
            target_answer = faq_df.iloc[20]['Answer']
        elif any(w in user_query for w in ["tech", "stack", "language", "php", "sql", "backend"]):
            target_answer = faq_df.iloc[9]['Answer']
        else:
            # Token matching loop as a safe backup mechanism
            max_matches = 0
            for index, row in faq_df.iterrows():
                q_text = str(row['Question'])
                matches = sum(1 for word in q_text.split() if len(word) > 3 and word in user_query)
                if matches > max_matches:
                    max_matches = matches
                    target_answer = row['Answer']

        # Final evaluation dispatch gate
        if target_answer:
            print(f"Bot: {target_answer}\n")
        else:
            print("Bot: I am sorry, I couldn't find a direct keyword match for that specific question in our database. Please rephrase your query or contact human operations.\n")

# Fire up the runtime chatbot loop
run_coretech_keyword_chatbot()
```
