```python
# Task 09: CoreTech Algorithmic Lead Scoring & Service Recommendation Engine (Upgraded)
import pandas as pd
import random

# Setting explicit seed for production reproducibility
random.seed(42)

# Calibrating realistic lead attributes as per evaluation guidelines
budgets = [800, 1500, 3500, 7500, 15000]
timelines = ["Immediate", "1 Month", "3 Months", "6 Months"]
sources = ["Website Form", "LinkedIn Outbound", "Cold Email", "Referral"]
company_sizes = ["1-10 (Startup)", "11-50 (SMB)", "51-200 (Medium Enterprise)", "200+ (Corporate)"]
urgencies = ["High", "Medium", "Low"]
engagement_levels = ["High Engagement", "Passive Interaction", "Cold Cold Outbound"] # New Attribute

leads_list = []
# Automatically generating 50 highly realistic client sample leads
for i in range(1, 51):
    leads_list.append({
        "Lead_ID": f"CTL-{i:03d}",
        "Client_Name": f"Client Venture {i}",
        "Budget_USD": random.choice(budgets),
        "Timeline": random.choice(timelines),
        "Source": random.choice(sources),
        "Company_Size": random.choice(company_sizes),
        "Urgency": random.choice(urgencies),
        "Engagement_Level": random.choice(engagement_levels)
    })

df_raw = pd.DataFrame(leads_list)
df_raw.to_csv('coretech_leads.csv', index=False)
print("1. Upgraded database 'coretech_leads.csv' successfully logged with engagement signals.")


# --- STEP 2: RE-ENGINEERED ALGORITHMIC SCORING ENGINE ---
leads_df = pd.read_csv('coretech_leads.csv')

def advanced_lead_scoring_logic(row):
    score = 0

    # A. Budget Signals Weight (Max: 35 Points)
    budget = row['Budget_USD']
    if budget >= 10000: score += 35
    elif budget >= 5000: score += 25
    elif budget >= 3000: score += 15
    elif budget >= 1500: score += 10
    else: score += 5

    # B. Urgency & Timeline Weights (Max: 25 Points)
    urgency = row['Urgency']
    timeline = row['Timeline']
    if urgency == "High" or timeline == "Immediate": score += 25
    elif urgency == "Medium" or timeline == "1 Month": score += 15
    else: score += 5

    # C. Company Size Signals (Max: 15 Points)
    size = row['Company_Size']
    if "Corporate" in size or "Medium" in size: score += 15
    elif "SMB" in size: score += 10
    else: score += 5

    # D. Engagement Level Weights - Explicitly Added (Max: 15 Points)
    engage = row['Engagement_Level']
    if engage == "High Engagement": score += 15
    elif engage == "Passive Interaction": score += 10
    else: score += 5

    # E. Lead Source Reliability Weights (Max: 10 Points)
    source = row['Source']
    if source == "Referral" or source == "Website Form": score += 10
    else: score += 5

    # F. Final Matrix Priority Segmentation Mapping
    if score >= 70: priority = "High Priority"
    elif score >= 45: priority = "Medium Priority"
    else: priority = "Low Priority"

    # G. Service Recommendation & Architecture Mapping Logic
    if budget >= 10000:
        service = "AI Automation Pipeline + Enterprise SQL Architecture"
        explanation = f"Enterprise-tier profile with elite high investment potential (${budget:,d} USD) and validated corporate scale."
    elif budget >= 3000:
        service = "Custom Full-Stack Web Application + UI/UX Optimization"
        explanation = f"Mid-tier engineering allocation optimized for professional scalable software workflows."
    elif budget >= 1500:
        service = "Responsive Frontend Development + Smart Chatbot Script"
        explanation = "Standard startup configuration leveraging vanilla frontend elements for client onboarding."
    else:
        service = "Digital Branding Suite & Interactive Figma Mockups"
        explanation = "Low baseline allocation suited for core asset visual layout branding deliverables."

    return pd.Series([score, priority, service, explanation])

# Run evaluation vectors across all 50 samples
leads_df[['Lead_Score', 'Priority_Label', 'Recommended_Service', 'Technical_Explanation']] = leads_df.apply(advanced_lead_scoring_logic, axis=1)
leads_df.to_csv('coretech_scored_leads.csv', index=False)
print("2. Core computational logic completed! 'coretech_scored_leads.csv' optimized.")
```
