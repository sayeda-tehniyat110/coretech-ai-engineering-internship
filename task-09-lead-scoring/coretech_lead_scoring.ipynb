```python
# Task 09: Combined Lead Generator and Scoring Engine Pipeline (Fixed String Formatting)
import pandas as pd
import random

# --- STEP 1: GENERATE DATASET ---
random.seed(42)

budgets = [800, 1500, 3500, 7500, 15000]
timelines = ["Immediate", "1 Month", "3 Months", "6 Months"]
sources = ["Website Form", "LinkedIn Outbound", "Cold Email", "Referral"]
company_sizes = ["1-10 (Startup)", "11-50 (SMB)", "51-200 (Medium Enterprise)", "200+ (Corporate)"]
urgencies = ["High", "Medium", "Low"]

leads_list = []
for i in range(1, 51):
    leads_list.append({
        "Lead_ID": f"CTL-{i:03d}",
        "Client_Name": f"Client Venture {i}",
        "Budget_USD": random.choice(budgets),
        "Timeline": random.choice(timelines),
        "Source": random.choice(sources),
        "Company_Size": random.choice(company_sizes),
        "Urgency": random.choice(urgencies)
    })

# Save to local virtual machine workspace
df_initial = pd.DataFrame(leads_list)
df_initial.to_csv('coretech_leads.csv', index=False)
print("1. 'coretech_leads.csv' successfully generated.")


# --- STEP 2: LOAD AND CALCULATE ---
leads_df = pd.read_csv('coretech_leads.csv')

def calculate_lead_score_and_route(row):
    score = 0

    # A. Budget Scoring (Max: 40)
    budget = row['Budget_USD']
    if budget >= 10000: score += 40
    elif budget >= 5000: score += 30
    elif budget >= 3000: score += 20
    elif budget >= 1500: score += 10
    else: score += 5

    # B. Urgency & Timeline (Max: 30)
    urgency = row['Urgency']
    timeline = row['Timeline']
    if urgency == "High" or timeline == "Immediate": score += 30
    elif urgency == "Medium" or timeline == "1 Month": score += 20
    else: score += 10

    # C. Company Size (Max: 15)
    size = row['Company_Size']
    if "Corporate" in size or "Medium" in size: score += 15
    elif "SMB" in size: score += 10
    else: score += 5

    # D. Lead Source (Max: 15)
    source = row['Source']
    if source == "Referral" or source == "Website Form": score += 15
    else: score += 10

    # E. Priority Matrix
    if score >= 75: priority = "High Priority"
    elif score >= 45: priority = "Medium Priority"
    else: priority = "Low Priority"

    # F. Service Routing (Fixed precision formatting syntax here)
    if budget >= 10000:
        service = "AI Automation Pipeline + Enterprise SQL Architecture"
        explanation = f"Large scale enterprise profile with elite high investment potential (${budget:,d} USD)."
    elif budget >= 3000:
        service = "Custom Full-Stack Web Application + UI/UX Optimization"
        explanation = f"Mid-tier investment optimized for professional deployment workflows."
    elif budget >= 1500:
        service = "Responsive Frontend Development + Smart Chatbot Script"
        explanation = "Standard startup configuration leveraging vanilla frontend components."
    else:
        service = "Digital Branding Suite & Interactive Figma Mockups"
        explanation = "Low baseline allocation suited for core layout branding deliverables."

    return pd.Series([score, priority, service, explanation])

# Run algorithmic loops
leads_df[['Lead_Score', 'Priority_Label', 'Recommended_Service', 'Technical_Explanation']] = leads_df.apply(calculate_lead_score_and_route, axis=1)

# Export final computed results
leads_df.to_csv('coretech_scored_leads.csv', index=False)
print("2. Lead Scoring calculations completed! 'coretech_scored_leads.csv' ready.")
```
