# CoreTech Internship Task:
**Prepared by:** AI Engineering Intern  
**Task Objective:** Testing how professional prompt structures improve AI code and content outputs compared to basic, lazy prompts.

## Category 1: Website Content (My Engineering Approach)

### Prompt 1.1: CoreTech Homepage Hero Headline
* **My First Try (Initial Prompt):**
   Write a hero headline for a software development agency web page.
* **What AI Gave Me (Expected Output):**
   "We build high-quality software applications for your business needs." (Too generic, sounds like every other company).
* **My Improved Version:**
   *Act as a tech startup copywriter. Write a powerful hero headline and subheadline for our agency 'CoreTech'. We specialize in custom full-stack web applications for scalable startups. Make it sound modern, tech-focused, and project an ambitious brand image.*
* **Why this works (My Explanation):**
   Giving it a role (startup copywriter) and mentioning our specific focus (full-stack web apps for startups) forces the AI to stop giving generic corporate fluff and write something sharp.



### Prompt 1.2: Portfolio Project Description (Alibaba Clone)
* **My First Try (Initial Prompt):**
   Write a short text describing an Alibaba clone project we built.
* **What AI Gave Me (Expected Output):**
   "We built an e-commerce application inspired by Alibaba featuring search and multiple grid layouts." (Sounds like a high school project).
* **My Improved Version:**
   *Draft a professional portfolio project breakdown for our 'Alibaba-inspired E-Commerce UI Clone'. Structure it with three headings: 1. The Challenge (Responsive CSS layout), 2. Our Backend & Frontend Stack (PHP, SQL database integration, dynamic product galleries), and 3. Final Outcome. Keep the tone highly technical.*
* **Why this works (My Explanation):**
   I explicitly forced the AI to use our engineering stack keywords (PHP, SQL, CSS Grid). This changes a simple description into a proper technical case study that looks good on GitHub.
  

### Prompt 1.3: CoreTech Services Page Layout
* **My First Try (Initial Prompt):**

For our three core services (Full-Stack Web Development, UI/UX Optimization, and Graphic Design/Branding), write a catchy benefit-driven title and 3 bullet points showing what the client gets. Keep the focus on how we help businesses grow, not just tech jargon.*

## Category 3: Client Emails

### Prompt 3.1: Replying to a New Client Inquiry
* **My Simple Try:**
   Email a potential client back who wants a web platform built.
* **What AI Gave Me:**
  "Thanks for contacting us. Let us know when you want to call to talk about your custom web platform."(Too short and not very professional).
* **My Better Prompt:**
   *Write a polite email reply to a new client who filled out our website form asking for a web application. Use placeholders like [Client Name]. Thank them for reaching out, and ask 3 simple questions about their project requirements, budget, and timeline. End with an invitation to jump on a quick call.*
* **Why this works:**
   By telling AI to ask 3 specific questions (requirements, budget, timeline), the email helps us get all the important project details before we start working.

### Prompt 3.2: Sending Figma Designs for Review
* **My Simple Try:**
  Tell the client the UI/UX mockups are finished in Figma.
* **What AI Gave Me:**
   "Hey, the mockups are ready on Figma. Look over them and send your review when you can." (Sounds too casual, like talking to a friend).
* **My Better Prompt:**
   *Write a professional project update email to a client. Tell them that our CoreTech team has finished the UI/UX high-fidelity designs in Figma. Give them clear steps on how to open the Figma link and view the screens. Ask them to share their feedback within 4 days so we can stay on schedule.*
* **Why this works:**
   Setting a deadline ("within 4 days") keeps the project on track and ensures the client knows their feedback is needed quickly to avoid delays.

### Prompt 3.3: Handling Out-of-Scope Requests (Scope Creep)
* **My Simple Try:**
   Reject a client request to add extra features for free.
* **What AI Gave Me:**
   "We cannot add these features because they were not part of the initial payment agreement." (Sounds very rude and harsh).
* **My Better Prompt:**
  *Act as a polite Project Manager. Write an email to a client who is asking for extra features that were not in our original agreement. First, praise their idea. Then, gently explain that this requires extra development hours. Offer them two options: add it as a paid extension now, or build it in Phase 2 after the main launch.*
* **Why this works:**
   Instead of saying a flat "No," this prompt helps us say no politely while offering helpful solutions, keeping the client relationship happy.



### Prompt 3.4: Friendly Payment Reminder Email
* **My Simple Try:**
   Send an invoice reminder email to a client who hasn't paid.
* **What AI Gave Me:**
   "Your invoice is overdue. Please pay it as soon as possible so we don't halt development." (Sounds panicked and aggressive).
* **My Better Prompt:**
   *Write a gentle but firm email reminding a client that their project invoice is 5 days overdue. Keep the tone completely objective and polite. Use placeholders like [Invoice Number]. Ask them to confirm if they have received the invoice or if they need any assistance with the payment process.*
* **Why this works:**
   It removes any awkward or angry language, making it look like a standard, professional automated system reminder.



### Prompt 3.5: Final Project Handover and Code Delivery
* **My Simple Try:**
   Send the final project code and documentation to the client.
* **What AI Gave Me:**
   "Attached is the source file zip archive. Thanks for working with our team on this project." (Too dry for such a huge milestone).
* **My Better Prompt:**
   *Write a comprehensive project closure email for CoreTech. Congratulate the client on their successful website launch. Provide clear details about what we are handing over (GitHub repository links, admin credentials, and documentation guide). At the end, mention that we also offer monthly maintenance services to help them long-term.*
* **Why this works:**
   It celebrates the launch, itemizes everything clearly so there's no confusion, and opens the door to upselling our monthly maintenance package.

* **My Simple Try:**
   What should a chatbot say when a server crashes?
* **What AI Gave Me:**
   "Error 500: The server is offline. Please refresh your browser window later." (Cold).
* **My Better Prompt:**
