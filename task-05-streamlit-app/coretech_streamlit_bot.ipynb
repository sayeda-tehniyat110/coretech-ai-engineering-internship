```python
%%writefile app.py
import streamlit as st
import pandas as pd
import io

st.set_page_config(
    page_title="CoreTech Web Assistant",
    page_icon="🤖",
    layout="centered"
)

st.markdown("""
<style>
.stApp { background-color: #0d0b18 !important; }
.stApp * { color: #ffffff !important; }
.stTextInput input {
    background-color: #1a1630 !important;
    color: #ffffff !important;
    border: 1px solid #5a3d99 !important;
}
.stButton button {
    background-color: #5a3d99 !important;
    color: #ffffff !important;
    border: none !important;
    border-radius: 8px !important;
    padding: 8px 20px !important;
}
</style>
""", unsafe_allow_html=True)

faq_data = """Question,Answer
"what services does coretech provide","CoreTech Innovation provides AI Automated Solutions, Full-Stack Web Development (PHP/SQL), UI/UX Optimization (Figma), and Graphic Branding Suite."
"where is coretech located","Our core corporate engineering hub is based in Sindh, Pakistan."
"who founded coretech","CoreTech was founded by a team of Computer Systems Engineering students specializing in scalable software and intelligent networks."
"what is the starting price for websites","A baseline Full-Stack Web Application with a chatbot starts at $1500 USD."
"how do i contact support or sales","Email us at operations@coretech.com or contact our support grid 24/7."
"what is your web development tech stack","We use PHP, SQL, HTML, CSS, and JavaScript."
"do you design logos","Yes, we offer logo design and full brand identity creation."
"do you use figma for ui design","Yes, we use Figma for wireframing and UI prototyping."
"how can i track my project progress","Clients get access to dedicated GitHub repositories for real-time updates."
"""

df_faq = pd.read_csv(io.StringIO(faq_data), dtype=str)

st.markdown("<h1 style='color:#b39ddb !important;font-weight:700;'>🤖 CoreTech Innovation</h1>", unsafe_allow_html=True)
st.markdown("<h3 style='color:#ffffff !important;'>Enterprise AI Onboarding Desk</h3>", unsafe_allow_html=True)
st.markdown("<p style='color:#cccccc !important;'>Ask me anything about CoreTech services or contact workflows.</p>", unsafe_allow_html=True)
st.markdown("---")

if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

for role, text in st.session_state.chat_history:
    if role == "user":
        st.markdown(f"""
        <div style='background:#1a1630;padding:12px;border-radius:10px;
        border:1px solid #5a3d99;margin-bottom:8px;'>
        <span style='color:#b39ddb;font-weight:bold;'>👤 You:</span>
        <span style='color:#ffffff;'> {text}</span>
        </div>""", unsafe_allow_html=True)
    else:
        st.markdown(f"""
        <div style='background:#16122c;padding:12px;border-radius:10px;
        border:1px solid #b39ddb;margin-bottom:8px;'>
        <span style='color:#7c4dff;font-weight:bold;'>🤖 CoreTech:</span>
        <span style='color:#ffffff;'> {text}</span>
        </div>""", unsafe_allow_html=True)

user_query = st.text_input("💬 Type your question here:", key="input_box", placeholder="e.g. What services do you offer?")
send = st.button("Send ➤")

if send and user_query.strip():
    clean_query = user_query.strip().lower().replace('?','').replace('.','')
    target_answer = None

    if any(w in clean_query for w in ["service", "provide", "offer", "do you do"]):
        target_answer = df_faq.iloc[0]['Answer']
    elif any(w in clean_query for w in ["where", "location", "located", "address", "pakistan"]):
        target_answer = df_faq.iloc[1]['Answer']
    elif any(w in clean_query for w in ["found", "who", "creator", "background"]):
        target_answer = df_faq.iloc[2]['Answer']
    elif any(w in clean_query for w in ["price", "cost", "investment", "fee", "charge"]):
        target_answer = df_faq.iloc[3]['Answer']
    elif any(w in clean_query for w in ["contact", "support", "email", "help", "sales"]):
        target_answer = df_faq.iloc[4]['Answer']
    elif any(w in clean_query for w in ["tech", "stack", "php", "sql", "backend"]):
        target_answer = df_faq.iloc[5]['Answer']
    elif any(w in clean_query for w in ["logo", "branding", "banner", "graphic"]):
        target_answer = df_faq.iloc[6]['Answer']
    elif any(w in clean_query for w in ["figma", "ui", "ux", "wireframe", "prototype"]):
        target_answer = df_faq.iloc[7]['Answer']
    elif any(w in clean_query for w in ["progress", "github", "commit", "repo"]):
        target_answer = df_faq.iloc[8]['Answer']
    else:
        max_matches = 0
        for _, row in df_faq.iterrows():
            q_text = str(row['Question'])
            matches = sum(1 for word in q_text.split() if len(word) > 3 and word in clean_query)
            if matches > max_matches:
                max_matches = matches
                target_answer = row['Answer']

    if not target_answer:
        target_answer = "I couldn't find a match. Please email us at operations@coretech.com."

    st.session_state.chat_history.append(("user", user_query))
    st.session_state.chat_history.append(("assistant", target_answer))
    st.rerun()
```
