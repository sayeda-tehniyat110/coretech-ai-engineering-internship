```python
# Task 11: Final Capstone Project - CoreTech Enterprise RAG Knowledge Assistant Ecosystem
!pip install -q sentence-transformers pandas

import numpy as np
import pandas as pd
import torch
import datetime
from sentence_transformers import SentenceTransformer

print(f"[{datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] Libraries loaded successfully.")

# 1. CoreTech Advanced Knowledge Base Dictionary
coretech_corpus = {
    "Company Profile": (
        "CoreTech Innovation is a premium computer systems engineering and software agency based in Sindh, Pakistan. "
        "Founded by engineering students, the firm bridges low-level hardware design, microprocessors (8085/8086), "
        "and numerical analysis logic with modern responsive full-stack applications, AI integrations, and advanced semantic data structures."
    ),
    "Services Portfolio": (
        "CoreTech Innovation delivers four primary production tracks: "
        "1. AI & NLP Automation (custom semantic search engines, LLM prompt engineering, automated customer routing workflows). "
        "2. Full-Stack Web Development (PHP backend frameworks, optimized SQL database clusters, dynamic content galleries). "
        "3. High-Fidelity UI/UX Prototyping (Figma wireframes, component design libraries, dynamic user journeys). "
        "4. Graphic Branding Suite (Custom corporate logos, brand guidelines, and social media layout kits)."
    ),
    "Engineering Process": (
        "Our rigorous engineering lifecycle is segmented into 4 agile phases: "
        "Phase 1: Discovery & Scoping (Analyzing requirements, setting API thresholds, and defining database schemas). "
        "Phase 2: Figma UI/UX Mockups (Building high-fidelity layout screens for client review). "
        "Phase 3: Active Coding Sprint (Writing clean documentation-first source code with transparent daily commits on GitHub branches). "
        "Phase 4: Triage & Production Deployment (Bug tracking, cloud configurations, and complete intellectual asset delivery)."
    ),
    "Pricing Framework": (
        "Project pricing is calculated dynamically based on engineering hours and backend architectural complexity. "
        "A standard Full-Stack Web Application combined with a rule-based AI automation script starts at a baseline sample investment of $1,500 USD. "
        "All projects require custom scope analysis, and absolute source code ownership is transferred to the client upon final milestone clearance."
    ),
    "Operational FAQs": (
        "Frequently Asked Questions (FAQs): "
        "Q1: Who owns the final code? A: The client receives absolute IP and ownership upon milestone clearance via GitHub. "
        "Q2: Can we audit ongoing coding tasks? A: Yes, clients receive direct repository access to track live git commits. "
        "Q3: Do you provide server deployment assistance? A: Yes, we handle product release deployment on cloud configurations."
    )
}

# Convert dictionary to a structural Pandas DataFrame and export as CSV for Capstone Audit
df_corpus = pd.DataFrame([{"Document_Title": k, "Content_Text": v} for k, v in coretech_corpus.items()])
df_corpus.to_csv("coretech_capstone_knowledge_base.csv", index=False)
print("-> Capstone audit database 'coretech_capstone_knowledge_base.csv' saved successfully.")
```

```python
# 2. Vectorization using Sentence-Transformers all-MiniLM-L6-v2
print("Loading semantic transformer mapping framework...")
embedding_engine = SentenceTransformer('all-MiniLM-L6-v2')

doc_titles = list(coretech_corpus.keys())
doc_contents = list(coretech_corpus.values())

# Compute vector spaces (384 dimensions)
indexed_embeddings = embedding_engine.encode(doc_contents, convert_to_tensor=True)
print(f"-> Successfully indexed {len(doc_titles)} enterprise records into vector spaces.")
```

```python
# 3. Capstone RAG Runtime Chat Assistant Interface Function
def run_coretech_capstone_assistant():
    print("\n" + "="*80)
    print("🤖 CORETECH ENTERPRISE CAPSTONE RAG KNOWLEDGE ASSISTANT INITIALIZED")
    print("Status: Active | Engine: all-MiniLM-L6-v2 | Scope: 5 Enterprise Assets")
    print("Type 'exit' or 'quit' to close the corporate terminal session.")
    print("="*80 + "\n")

    while True:
        query = input("Ask CoreTech Assistant: ").strip()

        # Termination conditions
        if query.lower() in ['exit', 'quit']:
            print("\n[SYSTEM LOG]: Capstone RAG interaction loop terminated gracefully. Logs closed. Goodbye!")
            break

        if not query:
            print("Assistant: Inquiry empty. Please rephrase your technical requirements.\n")
            continue

        # Semantic retrieval
        query_vector = embedding_engine.encode(query, convert_to_tensor=True)
        similarities = torch.nn.functional.cosine_similarity(query_vector, indexed_embeddings)

        # Extraction variables
        best_match_idx = torch.argmax(similarities).item()
        confidence_score = similarities[best_match_idx].item()

        source_doc = doc_titles[best_match_idx]
        context_text = doc_contents[best_match_idx]

        # Confidence boundaries threshold (0.25 standard limit)
        if confidence_score < 0.25:
            response = "I am sorry, I am unable to locate specific contextual logs matching those parameters in the CoreTech Innovation corporate database. Please contact human operations."
            source_doc = "None (Insufficient Structural Context)"
        else:
            # RAG synthesis: Generating response wrapped around the retrieved source text
            response = f"Based on verified CoreTech records [{source_doc}], {context_text}"

        # Display formatted enterprise output
        print("\n" + "-"*75)
        print(f"🔍 RETRIEVED SOURCE : {source_doc} (Score: {confidence_score:.4f})")
        print(f"🤖 GENERATED REPLY  :\n{response}")
        print("-"*75 + "\n")

# Initiate the interactive project execution loop
run_coretech_capstone_assistant()
```
